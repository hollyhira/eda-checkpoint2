## Visualization to get a general overview of cumulative cases vs deaths per 100,000 population for each region

**For analysis purposes, cumulative cases per 100,000 of the population will be used. This is because the comparison is across varying localities with different population sizes, so comparing per 100,000 establishes a baseline for better accuracy.**

```python
InteractiveShell.ast_node_interactivity = 'all'
```

```python
# pd.set_option('display.max_rows', None)
# covid_stats.groupby('WHO Region')['Name'].value_counts().to_frame()


covid_stats['WHO Region'].value_counts()

region_cases=covid_stats.groupby('WHO Region')['Cases - cumulative total per 100000 population'].mean().round(0).sort_values(ascending=False).to_frame()

region_deaths=covid_stats.groupby('WHO Region')['Deaths - cumulative total per 100000 population'].mean().round(0).sort_values(ascending=False).to_frame()

#comma separated figures
region_cases=region_cases.style.format('{:,.0f}')

region_deaths=region_deaths.style.format('{:,.0f}')

region_cases

region_deaths

```

```python
plt.style.use('seaborn-poster')
c =['green', 'yellow', 'orange', 'purple', 'blue','red', 'black']

ax=covid_stats.groupby('WHO Region')['Cases - cumulative total per 100000 population'].mean().sort_values(ascending=True).round(1).plot(kind='barh', stacked=True, figsize=(16, 14), fontsize=18, color=c)
ax.set_title('Average Cases by Region per 100,000 population', fontsize=22, weight='bold')
ax.set_xlabel('Cases per 100,000', weight='bold', fontsize=22)
ax.set_ylabel('WHO Region', weight='bold', fontsize=22)

ax.legend(loc=8, fontsize=22)


#remmove scientific notation on x axis
plt.ticklabel_format(axis='x', style='plain')

#Annotate plot with color conditions
for p in ax.patches:
    width = p.get_width()
    if width < 20:
        clr = 'red'
    else:
        clr = 'black'
    plt.text(5+p.get_width(), p.get_y()+0.55*p.get_height(),
             '{:,.0f}'.format(width),color=clr, weight='bold',fontsize=20,
             ha='left', va='baseline')
    
  # '{1.0f}'
```

```python
plt.style.use('seaborn-poster')
c =['green', 'yellow', 'orange', 'purple', 'blue','red', 'black']

ax=covid_stats.groupby('WHO Region')['Deaths - cumulative total per 100000 population'].mean().round(0).sort_values().plot(kind='barh', stacked=True, figsize=(16, 14), fontsize=18, color=c)
ax.set_title('Average Deaths by Region per 100,000 population', fontsize=22, weight='bold')
ax.set_xlabel('Deaths per 100,000', weight='bold', fontsize=22)
ax.set_ylabel('WHO Region', weight='bold', fontsize=22)
#ax.set_color_cycle(['green', 'yellow', 'orange', 'purple', 'blue','red', 'black'])

ax.legend(loc=8, fontsize=22)


#remmove scientific notation on x axis
plt.ticklabel_format(axis='x', style='plain')


#Annotate plot with color conditions
for p in ax.patches:
    width = p.get_width()
    if width < 20:
        clr = 'red'
    else:
        clr = 'black'
    plt.text(5+p.get_width(), p.get_y()+0.55*p.get_height(),
             '{:,.0f}'.format(width),color=clr, weight='bold',fontsize=20,
             ha='left', va='baseline')
```
